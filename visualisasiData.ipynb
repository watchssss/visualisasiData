### **Import Data**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "dataset.csv"
df = pd.read_csv(file_path)

df.head()
```

### **Cek Dataset Kosong**


```python
df.info()

df.isnull().sum()
```

### **Distribusi Jenis Konten (Movies vs TV Shows)**

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', palette='coolwarm')
plt.title("Distribusi Jenis Konten (Movies vs TV Shows)")
plt.xlabel("Jenis Konten")
plt.ylabel("Jumlah")
plt.show()
```

### **Tren Perilisan Konten per Tahun**

```python
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

plt.figure(figsize=(12, 6))
sns.histplot(df['release_year'], bins=30, kde=True, color="purple")
plt.title("Tren Perilisan Konten Netflix per Tahun")
plt.xlabel("Tahun Rilis")
plt.ylabel("Jumlah Konten")
plt.show()
```

### **Negara dengan Produksi Konten Terbanyak**

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("10 Negara dengan Produksi Konten Terbanyak di Netflix")
plt.xlabel("Jumlah Konten")
plt.ylabel("Negara")
plt.show()
```

### **Distribusi** **Rating**

```python
plt.figure(figsize=(10, 5))
sns.countplot(y=df['rating'], order=df['rating'].value_counts().index, palette='magma')
plt.title("Distribusi Rating Konten di Netflix")
plt.xlabel("Jumlah Konten")
plt.ylabel("Rating")
plt.show()
```

## **Insight Yang Didapat**

### 1. Netflix lebih banyak menyediakan film
Dari hasil visualisasi diatas dapat dilihat bahwa Netflix masih berfokus pada film sebagai konten utamanya dimana jumlah film 2 kali lebih besar dari TV shows

### 2. Perilisan Konten yang Drastis dari tahun ke tahun
Dari visualisasi tersebut terlihat bahwa jumlah konten yang dirilis meningkat drastis dalam beberapa tahun terakhir, terutama setelah tahun 2015, walaupun masih ada penurunan di tahun - tahun tertentu. Ini dapat disimpulkan bahwa Netflix secara agresif memproduksi konten terutama setelah tahun 2015, hal ini kemungkinan besar dipengaruhi oleh ekspansi global Netflix dalam produksi konten original.

### 3. Amerika Serikat Negara dengan Produksi Konten Terbanyak
Dominasi Amerika Serikat menunjukkan bahwa sebagian besar konten di Netflix berasal dari industri film dan TV Hollywood. selain US, India juga berada di posisi tinggi, yang menunjukkan bahwa Netflix juga serius dalam menargetkan pasar India dengan film Bollywood dan serial lokal. Negara-negara lain seperti Inggris dan Kanada juga berkontribusi cukup besar dalam produksi konten Netflix.

### 4. Konten Netflix mayoritas untuk ditonton oleh orang dewasa
Rating paling umum pada konten di Netflix adalah TV-MA (konten untuk dewasa), diikuti oleh TV-14 dan PG-13. Sehingga, platform ini lebih ditujukan untuk orang dewasa daripada anak - anak, walaupun ada beberapa konten dengan rating ramah keluarga (seperti PG dan G), sebagian besar kontennya lebih cocok untuk remaja dan dewasa.
